```python
# 1.) Capitalize()
# it help to Converts the first character to upper case.


msg = "hello, and welcome to scholarnuture."

x = msg.capitalize()

print (x)

# 2.) casefold()
# It helps to Converts string into lower case.

txt = "Hello, And Welcome To My World!"

x = txt.casefold()

print(x)

# 3.) count()
# It returns the number of times a specified value occurs in a string.

msg = "I love apples, apple are my favorite fruit"

x = msg.count("apple")

print(x)

# 4.) lower()
# Converts a string into lower case

txt = "Hello my FRIENDS"

x = txt.lower()

print(x)

# 5.) islower()
# Returns True if all characters in the string are lower case.

txt = "hello world!"

x = txt.islower()

print(x)

# 6.) isupper()
# Returns True if all characters in the string are upper case.

txt = "THIS IS NOW!"

x = txt.isupper()

print(x)

# 7.) isalpha()
# Checks if All Characters are Alphabets

name = "Monica"
print(name.isalpha())

# 8.) upper()
# Converts a string into upper case

txt = "Hello my friends"

x = txt.upper()

print(x)

# 9.) index()
# Searches the string for a specified value and returns the position of where it was found.

txt = "Hello, welcome to my world."

x = txt.index("welcome")

print(x)

# 9.) find()
# Searches the string for a specified value and returns the position of where it was found.

txt = "Hello, welcome to my world."

x = txt.find("welcome")

print(x)

# 10.) split()
# Splits the string at the specified separator, and returns a list.

txt = "welcome to the jungle"

x = txt.split()

print(x)

# 11.) startswith
# Returns true if the string starts with the specified value.

txt = "Hello, welcome to my world."

x = txt.startswith("Hello")

print(x)

# 12.) center()
# Returns a centered string

txt = "banana"

x = txt.center(20)

print(x)

# 13.) encode()
# Returns an encoded version of the string.

txt = "My name is St√•le"

x = txt.encode()

print(x)

# 14.) endswith()
# Returns true if the string ends with the specified value.

txt = "Hello, welcome to my world."

x = txt.endswith(".")

print(x)

# 15 .) format()
# Formats specified values in a string
```
